```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Tworzenie wykresu prostej

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]
```

```python
fig = plt.figure()

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Wykres funkcji y=x")

plt.plot(x, y, label='linear')
plt.legend()

plt.show()
```

### Wykres funkcji trygonometrycznych i nak≈Çadanie funkcji na jednym wykresie

```python
x = np.arange(0, 10, 0.2)
print("x = ", x)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.legend()

plt.show()
```

### Tworzenie subplot√≥w

```python
fig = plt.figure()
k, (ax1, ax2) = plt.subplots(2)

ax1.plot(x, np.sin(x), color='green', label='sinus')
ax2.plot(x, np.cos(x), color='orange', label='cosinus')

plt.show()
```

#### Lub drugim sposobem

```python
fig = plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x), color='green', label='sinus')

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x), color='orange', label='cosinus')

plt.show()
```

### Histogram z rozk≈Çadu normalnego

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target

df.head()
```

```python
cols = df.columns.values
combinations = []

for i in range(0, len(cols)):
    for j in range(i+1, len(cols)):
        combinations.append((cols[i], cols[j]))
```

```python
fig = plt.figure()

_, subplots = plt.subplots(len(combinations)//2, 2, figsize=(12,30))

subplots = subplots.flatten()
print(len(subplots))
for comb, sub in zip(combinations, subplots):
    sub.scatter(df[comb[0]], df[comb[1]], c=df['target'])
    sub.set_xlabel(comb[0])
    sub.set_ylabel(comb[1])

plt.show()
```
