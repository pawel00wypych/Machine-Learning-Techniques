## Operowanie na obrazie

```python
import numpy as np
import imageio.v2 as imageio
#from scipy.misc import imread, imresize
import matplotlib.pyplot as plt

img = imageio.imread('cat.jpg')
img.shape
```

```python
img_tinted = img * [0, 1, 1]

plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)

plt.imshow(np.uint8(img_tinted))
plt.show()
```

```python
img_resized = img_tinted.reshape(1024, 4000, 3)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)

plt.imshow(np.uint8(img_resized))
plt.show()
```

```python
img_small = img[500:1000,500:1000,:]
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)

plt.imshow(np.uint8(img_small))
plt.show()
```

```python
img_small.shape
```

```python
img_sm = img_small.reshape(250, 1000, 3)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img_small)

plt.subplot(1, 2, 2)

plt.imshow(np.uint8(img_sm))
plt.show()
```
